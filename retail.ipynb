```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, r2_score

# Example: X = features, y = target demand
tscv = TimeSeriesSplit(n_splits=5)

def train_model(X, y, model):
    rmse_scores = []
    mse_scores = []
    r2_scores = []

    for train_index, test_index in tscv.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        antilog_y_test = np.expm1(y_test)
        antilog_y_pred = np.expm1(y_pred)
        
        mse = mean_squared_error(antilog_y_test, antilog_y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(antilog_y_test, antilog_y_pred)
        rmse_scores.append(rmse)
        mse_scores.append(mse)
        r2_scores.append(r2)

    print("Walk-forward RMSE scores:", rmse_scores)
    print("Average RMSE:", np.mean(rmse_scores))
    print("\n")
    print("Walk-forward MSE scores:", mse_scores)
    print("Average MSE:", np.mean(mse_scores))
    print("\n")
    print("Walk-forward R2 scores:", r2_scores)
    print("Average R2:", np.mean(r2_scores))
```

```python
X_train = X_train.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

```python
from sklearn.metrics import  mean_squared_error, root_mean_squared_error, r2_score, mean_absolute_error

def evaluate_model(model, X_test, y_test):
    predictions = model.predict(X_test)

    antilog_y_test = np.expm1(y_test.clip(upper=20))
    antilog_predictions = np.expm1(np.clip(predictions, a_min=-10, a_max=20))

    mse = mean_squared_error(antilog_y_test, antilog_predictions)
    rmse = root_mean_squared_error(antilog_y_test, antilog_predictions)
    r2 = r2_score(antilog_y_test, antilog_predictions)
    mae = mean_absolute_error(antilog_y_test, antilog_predictions)

    return {
        'MSE': mse,
        'RMSE': rmse,
        'R2': r2,
        'MAE': mae
    }
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# loop over unique products
forecasts = {}
rmse_scores = []
for product in data["Product ID"].unique():
    product_data = data[data["Product ID"] == product].groupby("Date")["Demand"].sum()

    # ARIMA requires a time series index
    product_data = product_data.asfreq("D").fillna(0)  # daily freq, fill gaps
    
    train = product_data.iloc[:-30]  # all but last 30 days
    test = product_data.iloc[-30:]   # last 30 days (holdout for evaluation)

    # fit ARIMA
    model = ARIMA(train, order=(1,1,1)) 
    fitted = model.fit()

    # forecast next 30 days
    forecast = fitted.forecast(steps=len(test)).round().clip(lower=0)  # round demand up, no negatives
    forecasts[product] = forecast
    
    mae = mean_absolute_error(test, forecast)
    rmse = np.sqrt(mean_squared_error(test, forecast))
    rmse_scores.append(rmse)
    
    print(f"Product {product} - MAE: {mae:.2f}, RMSE: {rmse:.2f}")
print(f"Average RMSE across products: {np.mean(rmse_scores):.2f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin

class RetailPreprocessor(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X.copy()
        
        X['Date'] = pd.to_datetime(X['Date'])
        X['Year'] = X['Date'].dt.year
        X['Month'] = X['Date'].dt.month
        
        X['Epidemic_Category'] = X['Category'] + '_' + X['Epidemic'].astype(str)
        X['Season_Category'] = X['Seasonality'] + '_' + X['Category']
        X['Discount_Capped'] = X['Discount'].clip(upper=15)
        X['has_high_discount'] = (X['Discount'] > 15).astype(int)
        X['Order_Fufillment_Rate'] = np.where(
            X['Units Ordered'] > 0,
            X['Units Sold'] / X['Units Ordered'],
            0
        )
        X = X.drop(columns=['Date', 'Store ID', 'Product ID', 'Discount'], axis=1)

        # X["lag_1"] = X.groupby("Product ID")["Demand"].shift(1)
        # X["lag_7"] = X.groupby("Product ID")["Demand"].shift(7)
        
        # X["roll_mean_7"] = (
        #     X.groupby("Product ID")["Demand"]
        #         .shift(1)  # dont include today
        #         .rolling(7)
        #         .mean()
        # )

        # data["roll_std_7"] = (
        #     data.groupby("Product ID")["Demand"]
        #         .shift(1)
        #         .rolling(7)
        #         .std()
        # )

        # 1. Drop unnecessary columns
        # X = X.drop(["Date", "Store ID", "Product ID"], axis=1)

        # # 2. Feature Engineering
        # X['Epidemic_Category'] = X['Category'] + '_' + X['Epidemic'].astype(str)
        # X['Season_Category'] = X['Seasonality'] + '_' + X['Category']
        # X['Discount_Capped'] = X['Discount'].clip(upper=15)
        # X['has_high_discount'] = (X['Discount'] > 15).astype(int)
        # X['Order_Fufillment_Rate'] = np.where(
        #     X['Units Ordered'] > 0,
        #     X['Units Sold'] / X['Units Ordered'],
        #     0
        # )
        return X
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import xgboost as xgb

# Columns after feature engineering
numeric_features = ["Price", "Units Sold", "Units Ordered", 
                    "Competitor Pricing", "Inventory Level", "Discount_Capped", 
                    "has_high_discount", "Order_Fufillment_Rate", 'Year', 'Month']

categorical_features = ["Category", "Region", "Seasonality", "Epidemic_Category", 
                         "Season_Category", "Weather Condition"]

# OneHotEncode categorical features
categorical_transformer = OneHotEncoder(handle_unknown="ignore")

# ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("num", "passthrough", numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)


# Full pipeline
pipeline = Pipeline(steps=[
    ("feature_engineering", RetailPreprocessor()),
    ("encoding", preprocessor),
    ("model", rf)
])
```
